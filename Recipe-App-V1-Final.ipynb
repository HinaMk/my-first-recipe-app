```python
# repeat your "app" again but now also add a sort criteria
# Also, using calories and yield (number of servings) calculate the calories per serving and print that out as well


import requests
from pprint import pprint

URL = "https://api.edamam.com/api/recipes/v2"

def search_edamam_recipes3(query, mealType, calories, sortby): 
    params = {
        "q": query,
        "app_id": keys.app_id,
        "app_key": keys.key,
        "type": "any",
        "field": ["label", "calories", "images", "ingredientLines", "totalCO2Emissions", "yield", "totalTime"],
        "mealType": mealType,
        "calories": calories,
        "sort": sortby # my code for sortby
    }

    response = requests.get(URL, params=params)
    print(response.status_code)

    if response.status_code == 200:
        result = response.json() 
        recipes = result.get('hits', []) 
        print("Received {} recipes.".format(len(recipes))) # print how many recipes we received

        if len(recipes) > 0: # if we have at least 1 recipe
            recipes.sort(key=lambda x: x['recipe'][sortby]) # Sort the recipes based on the specified sort criteria

            return recipes
    return None

def calculate_calories_per_serving(recipe): # calculate calories per serving
    calories = recipe['recipe']['calories'] # get the calories
    yield_value = recipe['recipe']['yield'] # get the yield (number of servings)
    if calories is not None and yield_value is not None and yield_value > 0: # if calories and yield_value are not None and yield_value is greater than 0
        calories_per_serving = calories / yield_value # calculate calories per serving
        return calories_per_serving # return calories per serving
    else:
        return None

query = "chicken"
mealType = "dinner"
sortby = "calories" 

sorted_recipes = search_edamam_recipes3(query, mealType, calories, sortby) 

if sorted_recipes is not None:
    for recipe in sorted_recipes: # loop over the recipes
        label = recipe['recipe']['label'] # get the recipe name
        calories = recipe['recipe']['calories'] # get the calories
        ingredients = recipe['recipe']['ingredientLines'] # get the ingredients
        image = recipe['recipe'].get('image', 'No image available') # get the image, if none, return 'No image available'

        calories_per_serving = calculate_calories_per_serving(recipe) # calculate calories per serving

        print(f"Recipe name: {label}") # print the recipe name
        print(f"Calories: {calories} kcal") # print the calories
        print(f"Calories per serving: {calories_per_serving} kcal") # print the calories per serving
        print("Ingredients:") # print "Ingredients:"
        for ingredient in ingredients: # loop over the ingredients
            print(f"- {ingredient}") # print the ingredient
        print(f"Image: {image}") # print the image
        print("\n") # print a new line
else:
    print("No recipes found")
```
